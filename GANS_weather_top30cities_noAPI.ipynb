```python
def forecast_n(cities):
    appid = #"open-weather-API-key"
    exclude = ["minutely","hourly","alerts"]
    city = []
    temp = []
    temp_felt = []
    humidity = []
    weather = []
    weather_det = []
    wind = []
    date = []
    for i in range(len(cities)):
        Lat = cities.Lat_dec[i]
        Long = cities.Long_dec[i]
        url = f"http://api.openweathermap.org/data/2.5/forecast?lat={Lat}&lon={Long}&exclude={exclude}&appid={appid}&units=metric"
        response = requests.get(url)
        wd = response.json()
        try:
            for w in wd['list']:
                if w.get('dt_txt').find("12:00:00")>0:
                    city.append(cities.City[i])
                    temp.append(w.get('main').get('temp'))
                    temp_felt.append(w.get('main').get('feels_like'))
                    humidity.append(w.get('main').get('humidity'))
                    weather.append(w.get('weather')[0].get('main'))
                    weather_det.append(w.get('weather')[0].get('description'))
                    wind.append(w.get('wind').get('speed'))
                    date.append(w.get('dt_txt').split(" ")[0])
                else:
                    continue
        except:
            continue
    weather_df = pd.DataFrame({
        "City": city,
        "Temperature": temp,
        "Feels_like": temp_felt,
        "Humidity": humidity,
        "Weather": weather,
        "Weather_Details": weather_det,
        "Windspeed": wind,
        "Date": date
        })
    weather_df["Date"] = pd.to_datetime(weather_df.Date) # not tested so far
    forecast = weather_df.set_index("City").drop_duplicates()
    forecast = forecast.reset_index() # not tested so far
    return forecast
```

```python
cities = EU.head(30)
```

```python
result = forecast_n(cities)
result
```

```python
result = result.reset_index()
```
