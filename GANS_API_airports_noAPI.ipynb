```python
filepath = r"C:\Users\danie\Desktop\WBS.DataScience_Bootcamp\Week_6_Webscraping_API\EU_cities.csv"
EU_cities = pd.read_csv(filepath, sep = ",")
EU = EU_cities.copy()
```

```python
EU_top_30 = EU.head(30)
EU_top_30
```

```python
latitudes = []
for i in EU_top_30.Lat_dec:
    latitudes.append(i)
latitudes
```

```python
longitudes = []
for i in EU_top_30.Long_dec:
    longitudes.append(i)
longitudes
```

```python
def icao_airport_codes(latitudes, longitudes):

  #assert len(latitudes) == len(longitudes)
  
  list_for_df = []

  for i in range(len(latitudes)):
    try:
      url = f"https://aerodatabox.p.rapidapi.com/airports/search/location/{latitudes[i]}/{longitudes[i]}/km/50/16"

      querystring = {"withFlightInfoOnly":"true"}

      headers = {
	  "X-RapidAPI-Key": #"aerodatabox-API-key",
	  "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}

      response = requests.request("GET", url, headers=headers, params=querystring)
      city = []
      AP = pd.json_normalize(response.json()["items"])
      for n in range(len(AP)):
          city.append(EU_top_30.City[i])
      AP.insert(loc = 0,
          	column = 'City',
          	value = city)
      list_for_df.append(AP)
    except:
        continue
  return pd.concat(list_for_df, ignore_index=True)
```

```python
Top_30_city_airports = icao_airport_codes(latitudes, longitudes)
```

```python
Top_30_city_airports
```

```python
City_ICAO = Top_30_city_airports.iloc[:, [0,1]]
City_ICAO
```

```python
City_ICAO.to_csv("Top30_cities_airports_50km.csv", index=False)
```
